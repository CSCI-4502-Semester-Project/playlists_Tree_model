```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
bayes = pd.read_csv('./trial_logs/NaiveBayes.csv').rename(columns={'ElapsedTime':'ElapsedTime (s)', ' Score':'Score', ' FitTime':'FitTime (s)', ' AvgBranchTime':'AvgBranchTime (s)', ' AvgConf':'AvgConf'})
neural = pd.read_csv('./trial_logs/NeuralNet.csv').rename(columns={'ElapsedTime':'ElapsedTime (s)', ' Score':'Score', ' FitTime':'FitTime (s)', ' AvgBranchTime':'AvgBranchTime (s)', ' AvgConf':'AvgConf'})
forest = pd.read_csv('./trial_logs/RandomForest.csv').rename(columns={'ElapsedTime':'ElapsedTime (s)', ' Score':'Score', ' FitTime':'FitTime (s)', ' AvgBranchTime':'AvgBranchTime (s)', ' AvgConf':'AvgConf'})
svm = pd.read_csv('./trial_logs/SVM.csv').rename(columns={'ElapsedTime':'ElapsedTime (s)', ' Score':'Score', ' FitTime':'FitTime (s)', ' AvgBranchTime':'AvgBranchTime (s)', ' AvgConf':'AvgConf'})

bayes['AvgConf'] = (bayes['AvgConf'] - bayes['AvgConf'].min()) / bayes['AvgConf'].max()
neural['AvgConf'] = (neural['AvgConf'] - neural['AvgConf'].min()) / neural['AvgConf'].max()
forest['AvgConf'] = (forest['AvgConf'] - forest['AvgConf'].min()) / forest['AvgConf'].max()
svm['AvgConf'] = (svm['AvgConf'] - svm['AvgConf'].min()) / svm['AvgConf'].max()
```

```
fig, axes = plt.subplots(1, 5, figsize=(25, 5)) # create figure and axes

for i,el in enumerate(list(bayes.columns.values)[0:5]):
    a = bayes.hist(el, ax=axes.flatten()[i])

plt.tight_layout(pad=50)

plt.show()

print('\nMean') 
print(bayes.mean())
print('\nSTD')
print(bayes.std())
```
